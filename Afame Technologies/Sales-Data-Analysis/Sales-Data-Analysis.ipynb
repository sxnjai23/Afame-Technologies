```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_excel("ECOMM_DATA.xlsx")
```

```python
total_sales = sales_data['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data['Ship Date'] = pd.to_datetime(sales_data['Ship Date'])
```

```python
monthly_sales = sales_data.groupby(sales_data['Order Date'].dt.to_period('M'))['Sales'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Year-Month')
plt.ylabel('Sales Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

best_selling_products = sales_data.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10)
print("Best-Selling Products:\n", best_selling_products)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=best_selling_products.values, y=best_selling_products.index, palette='viridis')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Quantity Sold')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```
